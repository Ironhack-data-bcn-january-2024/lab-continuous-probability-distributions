## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
import numpy as np

x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here

import matplotlib.pyplot as plt
from scipy.stats import uniform

x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)

def uniform_numbers(bottom, ceiling, count):
    x = uniform.rvs(size=count)
    randoms = bottom + (ceiling - bottom) * x
    return randoms

set1 = uniform_numbers(bottom=10, ceiling=15, count=100)

plt.subplot(1, 2, 1)
plt.hist(set1, bins=10, color='blue', edgecolor='black')
plt.title('Uniform Distribution 1')
plt.xlabel('Value')
plt.ylabel('Count')

set2 = uniform_numbers(bottom=10, ceiling=60, count=1000)

plt.subplot(1, 2, 2)
plt.hist(set2, bins=10, color='red', edgecolor='black')
plt.title('Uniform Distribution 2')
plt.xlabel('Value')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
# your code here

from scipy.stats import norm

def generate_normal_sample(mean, stdev, size):
    my_normal = norm(mean, stdev)
    sample = my_normal.rvs(size)
    
    return sample

sample = generate_normal_sample(10, 1, 1000)
sample2 = generate_normal_sample(10, 50, 1000)

plt.subplot(1, 2, 1)
plt.hist(sample, bins=30, color='blue', edgecolor='black')
plt.title('Normal Distribution 1')
plt.xlabel('Value')
plt.ylabel('Density')

plt.subplot(1, 2, 2)
plt.hist(sample2, bins=30, color='red', edgecolor='black')
plt.title('Normal Distribution 2')
plt.xlabel('Value')
plt.ylabel('Density')

plt.tight_layout()
plt.show()

```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 2 and 200. 

1. Use the function to generate two number sequences with the size of 1000 for both.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here

def exponential_numbers(mean, size):
    return np.random.exponential(scale=mean, size=size)

sequence1 = exponential_numbers(2, 1000)
sequence2 = exponential_numbers(200, 1000)

plt.subplot(1, 2, 1)
plt.hist(sequence1, bins=50, color='blue', edgecolor='black', alpha=0.7, label='Mean=2')
plt.subplot(1, 2, 2)
plt.hist(sequence2, bins=50, color='red', edgecolor='black', alpha=0.7, label='Mean=200')
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
